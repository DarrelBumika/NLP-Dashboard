```python
import pickle
import re
import pandas as pd
```

```python
model = pickle.load(open('models/models-nb.p', 'rb'))
feature_bow = pickle.load(open('models/feature-bow.p', 'rb'))
```

```python
def predict_sentiment(sent):
    sent = feature_bow.transform([sent])
    pred = model.predict(sent)
    return pred[0]
```

```python
data = pd.read_csv('tweets-data/cleaned-data.csv')
```

```python
data['sentiment'] = data['Tweet'].apply(predict_sentiment)
data.head(10)
```

```python
data.to_csv('tweets-data/sentiment-data.csv', index=False)
```
