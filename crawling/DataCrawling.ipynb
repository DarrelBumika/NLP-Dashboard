```python
import pandas as pd

file1_path = f"tweets-data/raw-data-september.csv"
file2_path = f"tweets-data/raw-data-october.csv"

data1 = pd.read_csv(file1_path)
data2 = pd.read_csv(file2_path)

data = pd.concat([data1, data2], ignore_index=True)
data.head()
```

```python
# Cek jumlah data yang didapatkan

num_tweets = len(data)
print(f"Jumlah tweet dalam dataframe adalah {num_tweets}.")
```

```python
data.to_csv("tweets-data/raw-data.csv", index=False)
```
